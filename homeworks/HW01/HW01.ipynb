### Примеры создания массивов

```python
# импорт библиотеки
import numpy as np
```

```python
# создание массива Numpy с помощью списка
np.array([[14, -2, 11],
          [-18, 0, 0],
          [0.24, 11, -0.0014],
          [546, 546, 34343]], 
          dtype=np.float32)
```

```python
# создание массива целых чисел Numpy с помощью списка
np.array([[14, -2, 11],
          [-18, 0, 0],
          [0.24, 11, -0.0014],
          [546, 546, 34343]], 
          dtype=np.int32)
```

```python
# создание единичной матрицы 4x4
np.eye(4, 4)
```

```python
# создание многомерного Numpy массива нулей 2x11x5
np.zeros(shape=(2, 11, 5), dtype=np.int8)
```

```python
# создание многомерного Numpy массива единиц 2x3x4x5
np.ones(shape=(2, 3, 4, 5))
```

```python
# создание Numpy массива в виде последовательности чисел с плавающей точкой
# от -10 включительно до 52 с шагом 3.5
np.arange(-10, 52, 3.5, dtype=np.float16)
```

```python
np.sin(np.empty(shape=(3, 6), dtype=np.double)).round(4)
```

### Размерность Numpy массивов
(некоторые типы данных уже были рассмотренны при создании массивов)

```python
# создадим Numpy массив в виде первых 100 натуральных чисел и узнаем его размерность
my_pretty_array = np.arange(1, 101)

print(f"Созданный массив: \n {my_pretty_array} \n")
print(f"Размерность массива: {my_pretty_array.shape}")
print(f"Кол-во элементов массива {len(my_pretty_array)}")
```

```python
# рассмотрим новые массивы, полученные изменением размерности исходного
print(f"Массив из 5 строк и 20 столбцов: \n {my_pretty_array.reshape((5, 20))} \n")
print(f"Размерность: {my_pretty_array.reshape((5, 20)).shape}")
```

```python
print(f"3D массив из 5 матриц размера 5x4: \n {my_pretty_array.reshape((5, 5, 4))} \n")

print(f"Размерность: {my_pretty_array.reshape((5, 5, 4)).shape}")
```

```python
# Рассмотрим массивы с N измерениями, где N-1 измерений заданы, 
# а одно определяется автоматически с помощью -1
print(f"Массив размерности {my_pretty_array.reshape((-1, 1)).shape}: \n {my_pretty_array.reshape((-1, 1))} \n")
```

```python
print(f"Массив размерности {my_pretty_array.reshape((10, -1, 10)).shape}: \n {my_pretty_array.reshape((10, -1, 10))} \n")
```

```python
print(f"Массив размерности {my_pretty_array.reshape((1, -1, 5)).shape}: \n {my_pretty_array.reshape((1, -1, 5))} \n")
```

```python
try:
    print(f"Пример использования более одной неизвестной размерности: \
      \n {my_pretty_array.reshape((-1, -1, 5))}")
except ValueError as e:
    print(e.args)

try:
    print(f"Пример невозможного изменения размерности: \
      \n {my_pretty_array.reshape((5, 5, 5))}")
except ValueError as e:
    print(e.args)
```

### Примеры индексирования и срезов

```python
# Отдельные элементы и срезы будем брать из my_pretty_array
# с изменённой размерностью (1, 20, 5)
my_pretty_array = my_pretty_array.reshape((1, -1, 5))
print(my_pretty_array.dtype, "\n", my_pretty_array)
```

```python
print(f"Первый элемент последней строки: {my_pretty_array[0][-1][0]}")
print(f"Первая строка: {my_pretty_array[0, 0]}")
print(f"Второй столбец с конца: {my_pretty_array[0, :, -2]}")
print(f"Медиана десятой строки: {my_pretty_array[0, 9, my_pretty_array.shape[-1]//2]}")
print(f"Трейий и пятый элементы каждой чётной строки: \n {my_pretty_array[0, ::2, 2::2]}")
```

```python
try:
    print(f"Пример некорректного индекса: \
      \n {my_pretty_array[0, 0.5]}")
except IndexError as e:
    print(e.args)
```

### Примеры векторных операций и базовых статистик

```python
# в качестве массива для преобразования возьмём срез my_pretty_array
my_new_pretty_array = my_pretty_array[0, ::3, ::2]
my_new_pretty_array
```

```python
# результат применения тригонометрических функций
print(f"Синус: \n {np.sin(my_new_pretty_array)}")
print(f"Гиперболический тангенс: \n {np.tanh(my_new_pretty_array)}")
print(f"Гиперболический арксинус: \n {np.arcsinh(my_new_pretty_array)}")
print(f"Арккосинус: \n {np.arccosh(my_new_pretty_array)}")
```

```python
# базовые статистики
print(f"Среднее каждой строки: \n {np.mean(my_new_pretty_array, axis=1)}")
print(f"Медиана каждой строки: \n {np.median(my_new_pretty_array, axis=1)}")
print(f"Стандартное отклонение для каждого столбца: \n {np.std(my_new_pretty_array, axis=0)}")
print(f"Сумма по каждой строке: \n {np.sum(my_new_pretty_array, axis=1)}")
print(f"Сумма по всему массиву: \n {np.sum(my_new_pretty_array)}")
print(f"Максимум по каждой строке: \n {np.max(my_new_pretty_array, axis=1)}")
```

```python
# некоторые операции с массивами
print(f"Сложение (поэлементно): \n {my_new_pretty_array + np.sin(my_new_pretty_array)}\n")
print(f"Умножение (поэлементно): \n {my_new_pretty_array * np.pow(my_new_pretty_array, 2)}\n")
print(f"Объединение по столбцам: \n {np.hstack([my_new_pretty_array, -0.25*my_new_pretty_array])}\n")
print(f"Объединение по строкам: \n {np.vstack([np.sin(my_new_pretty_array), np.pow(my_new_pretty_array, 2)]).round(4)}\n")
print(f"Умножение (векторное): \n {my_new_pretty_array @ my_new_pretty_array.reshape(3, -1)}")
```

```python
try:
    print(f"Пример некорректного векторного умножения (несоместимые размерности): \
      \n {my_new_pretty_array @ my_new_pretty_array}")
except ValueError as e:
    print(e.args)
```
